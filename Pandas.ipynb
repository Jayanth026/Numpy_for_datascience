**Series**

`In Pandas Library -> Series is a Built_in_function is used to represent only 1 column data `


```python
import numpy as np
import pandas as pd
```

```python
age = np.array([4,10,15,20,25]) # this is for dimensions

a = pd.Series(age) # giving 1D data to pandas series function

a
```

```python
age = np.array([4,10,15,20,25]) # this is for dimensions

a = pd.Series(age,index = ['row1','row2','row3','row4','row5']) # giving 1D data to pandas series function

a
```

```python
a['row3']
```

```python
# we can also use arange

a = np.arange(11,20,2)


b = pd.Series(a)

b
```

```python
b[3]
```
